```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
from scipy.stats import skew, kurtosis 

# List of CSV file paths of 2 years of historical data for 3 cryptocurrencies
csv_files = ['Bitcoin_cash.csv', 'Ethereum_classic.csv', 'Solana.csv']

# Dictionary to hold 'change' columns from each DataFrame
change_columns = {}

# Read each CSV file into a dataframe and store in list
for file in csv_files:

    # Step 1: Read the CSV into a DataFrame
    df = pd.read_csv(file)

    # Step 2: Convert the 'date' column to datetime format
    df['date'] = pd.to_datetime(df['date'])

    # Step 3: Sort the DataFrame by the 'date' column, oldest first
    df_sorted = df.sort_values(by='date')

    # Display the sorted DataFrame
    print(f"{file}")
    print(f"Top  5: {df_sorted.head()}")
    print("")     
    print(f"Average Return: {df_sorted['change']}")
    
    # Step 4: Prepare the 'change' column: remove percentage signs and convert to float
    df_sorted['change'] = df_sorted['change'].str.replace('%', '').astype(float) / 100

    # Step 5: Calculate the required statistics on the 'change' column
    std_dev = df_sorted['change'].std()
    skewness = df_sorted['change'].skew()
    kurt = df_sorted['change'].kurt()

    # Print the statistics
    print(f"Standard Deviation: {std_dev}")
    print(f"Skewness: {skewness}")
    print(f"Kurtosis: {kurt}")
    
    # Step 6: Use the file name without '.csv' as the key and store the 'change' column
    dataframe_key = file.replace('.csv', '')
    change_columns[dataframe_key] = df_sorted['change']
    
    print("-----------------------------------------")
   
# Step 7: After processing all files, create a DataFrame from the 'change' columns for correlation and covariance calculation
combined_change_df = pd.DataFrame(change_columns)

print("Combined DataFrame: {}".format(combined_change_df))

# Step 8: Calculate correlation matrix
correlation_matrix = combined_change_df.corr()

# Step 9: Calculate covariance matrix
covariance_matrix = combined_change_df.cov()

# Step 10: Optionally display the matrices
print("Correlation Matrix:")
print(correlation_matrix)
print("\nCovariance Matrix:")
print(covariance_matrix)   
```

# Plotting Output of Matrices

```python
import plotly.graph_objects as go

# Assuming 'correlation_matrix' and 'covariance_matrix' are already computed

# Plot Correlation Matrix
fig_corr = go.Figure(data=go.Heatmap(
    z=correlation_matrix.values,
    x=correlation_matrix.columns,
    y=correlation_matrix.index,
    colorscale='Viridis',  # You can change the colorscale as needed
    colorbar=dict(title='Correlation')
))
fig_corr.update_layout(title='Correlation Matrix', xaxis_title='Crypto classes', yaxis_title='Crypto classes')
fig_corr.show()

# Plot Covariance Matrix
fig_cov = go.Figure(data=go.Heatmap(
    z=covariance_matrix.values,
    x=covariance_matrix.columns,
    y=covariance_matrix.index,
    colorscale='Viridis',  # You can change the colorscale as needed
    colorbar=dict(title='Covariance')
))
fig_cov.update_layout(title='Covariance Matrix', xaxis_title='Crypto classes', yaxis_title='Crypto classes')
fig_cov.show()
```

### Shorting Cryptocurrency Assets

Short selling and credit risk are important considerations in cryptocurrency portfolios, similar to traditional financial markets but with distinct characteristics due to the digital and decentralized nature of crypto assets.

Cryptocurrencies can be sold short, allowing investors to profit from declining prices. The process involves borrowing the cryptocurrency, selling it at the current market price, and then buying it back later at a lower price to return to the lender, keeping the price difference as profit. This practice is facilitated by exchanges and special platforms that support crypto short selling.

**Mechanics of Short Selling:**

1. **Borrowing**: Obtain the cryptocurrency from a lender, which could be an exchange or another trader.
2. **Selling**: Sell the borrowed assets at the current market price.
3. **Buying Back**: Repurchase the assets at a lower price if the market moves as expected.
4. **Returning**: Give back the borrowed amount to the lender, pocketing the difference minus any fees.

Short selling in crypto carries risks such as high volatility, regulatory changes, and the potential for short squeezes, which can lead to rapid price increases and significant losses for short sellers.

### Credit Risk in Cryptocurrency Portfolios

Credit risk in crypto portfolios primarily revolves around counterparty risk and the inherent risks in lending and borrowing within the cryptocurrency space.

**Forms of Credit Risk:**

1. **Counterparty Risk**: This arises when an entity (e.g., an exchange or lending platform) fails to fulfill its financial obligations, which can be due to operational failures, security breaches, or insolvency.

2. **Lending/Borrowing Risks**: Engaging in crypto lending, especially through decentralized finance (DeFi) platforms, introduces the risk of borrower default, which can lead to losses.

3. **Smart Contract Risk**: Investments in DeFi or other smart contract-based instruments carry the risk of contract failures, bugs, or exploits, potentially leading to financial losses.

4. **Regulatory Risk**: Changes in crypto regulations can impact the value of assets or the ability of entities to meet their obligations, indirectly contributing to credit risk.

Investors in cryptocurrency need to be aware of these unique risks and consider the volatile and evolving nature of the crypto market. While opportunities for profit exist, such as through short selling, the associated risks, including credit risk, require careful management and a deep understanding of the crypto ecosystem.
